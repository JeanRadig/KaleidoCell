# Tutorial 04: assess the convergence of kaleidoCell

We show here how to check whether kaleidoCell has correctly converged. We remind that there are two convergence parameters that can be looked at. The number of initialisations (n_initializations), which is the amount of random restart of the matrices $W$ and $H$ at each iteration; and the number of iterations, which are the amount of steps over which $W$ and $H$ are optimised. Hence, schematically, you are going to create as many runs as you have initialisations and you let $W$ and $H$ converge for as many steps as max_iterations. For the results obtained in our paper, we saw that we only gained marginal benefits of running for several initialisations and for much longer than 100 iterations, but nonetheless it is of importance for the user to be able to check whether convergence has indeed been reached.

## 0 · Imports and data preparation

```python
import kaleidocell
import scanpy as sc
import pandas as pd

adata = sc.read_h5ad('../data/petersims_example.h5ad')
```

## 1 · Plot convergence plots - 1 initialization

```python
results_nmf, nmf_convergence = kaleidocell.multi_sample_nmf(
    adata,
    batch_key='Patients',
    test_ranks=[4, 5, 6, 7, 8, 9],
    n_initializations=1,
    max_iterations=100
)

## Runtime: 3.4s 
```

```python
kaleidocell.plot_convergence_plots(nmf_convergence)
```

## 2 · Plot convergence plots - 10 initializations

```python
results_nmf, nmf_convergence = kaleidocell.multi_sample_nmf(
    adata,
    batch_key='Patients',
    test_ranks=[4, 5, 6, 7, 8, 9],
    n_initializations=10,
    max_iterations=100
)

## Runtime: 32.3s
```

```python
kaleidocell.plot_convergence_plots(nmf_convergence)
```

```python
results_nmf, nmf_convergence = kaleidocell.multi_sample_nmf(
    adata,
    batch_key='Patients',
    test_ranks=[4, 5, 6, 7, 8, 9],
    n_initializations=1,
    max_iterations=1000
)

## Runtime: 24.9s
```
